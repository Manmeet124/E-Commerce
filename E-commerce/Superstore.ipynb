```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
dataframe = pd.read_csv("C:/Users/Lenovo/Desktop/E commerce.csv", encoding='latin1')
print(dataframe)
```

```python
#SUB_CATEGORY
dataframe['Sub-Category'].value_counts()
```

```python
#COUNT OF CATEGORY
plt.figure(figsize=(5,4))
sns.countplot(x="Category", data= dataframe)
plt.title("Count Of Category")
plt.savefig("Count Of Category.jpg")
plt.show()
    
```

```python
#SHIPPING MODE
dataframe['Ship Mode'].value_counts()
```

```python
x= dataframe['Ship Mode'].value_counts().index
y=dataframe['Ship Mode'].value_counts().values
```

```python
plt.figure(figsize=(5,4))
plt.pie(y,labels=x,startangle= 60,autopct="%0.2f%%")
plt.legend(loc=2)
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='Ship Mode',data=dataframe,hue="Category")
plt.show()
```

```python
#SEGMENT
plt.figure(figsize=(10,6))
sns.countplot(x='Category',data= dataframe[dataframe['Category']=='Office Supplies'], hue="Sub-Category")
plt.show()
```

```python
#SEGMENT
plt.figure(figsize=(5,4))
sns.countplot(x='Segment',data= dataframe)
plt.show()
```

```python
sns.barplot(x='Category',y="Profit", data=dataframe, estimator="sum")
plt.show()
```
